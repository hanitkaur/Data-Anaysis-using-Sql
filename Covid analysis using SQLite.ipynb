# ANALYZING DEATHS AND CASES

```python
import sqlite3
con = sqlite3.connect('covid')
cur = con.cursor()
r=cur.execute("select * from covid_death;")
l=[]
for i in r.description:
    l.append(i[0])
df=pd.DataFrame(r,columns=l)
df
```

```python
# get list of columns

r=cur.execute("select * from covid_death;")
l=[]
for i in r.description:
    l.append(i[0])
l
```

```python
# get the list of countries with maximum cases
print(":::Top 15 Countries with Highest Cases:::")
print("-----------------------------------------")

q1="select location, max(total_cases) as t from covid_death where continent<>'' group by location having max(total_cases)<>'' order by t desc limit 15;"
r=cur.execute(q1)
df1=pd.DataFrame(r,columns=["Country" ,"Total Cases"])
df1
```

```python
# get the name ofcountries that were having less cases
print(":::Top 15 Countries with Minimum Cases:::")
print("-----------------------------------------------")
q1="select location,sum(new_cases) as t from covid_death where continent<>'' group by location order by t limit 15;"
r=cur.execute(q1)
df1=pd.DataFrame(r,columns=["Country" ,"Total Cases"])
df1 
```

```python
# get percentage of population affected for top 15 countries
print(":::Percentage of Population Affected of top 15 countries with highest number of cases :::")
print("-----------------------------------------------------------------------------------------")
q1="select location,max(total_cases) as t, population, ((max(total_cases))*100.0)/population as pp from covid_death where continent<>'' group by location having t<>'' order by t desc limit 15;"
r=cur.execute(q1) 
df1=pd.DataFrame(r,columns=["Country" ,"Total Cases","Population","Percentage of Population Affected"])
df1 
```

```python
# get death ratio of top 15 countries with maxium number of cases
print(":::Death Percentage of countries with Maximum number of cases:::")
print("-----------------------------------------------")
q1="select location, max(total_cases) as t, sum(new_deaths) as td,(sum(new_deaths)*100.0)/(max(total_cases)) as pd from covid_death where continent<>'' group by location having t<>'' order by t desc limit 15;"
r=cur.execute(q1) 
df1=pd.DataFrame(r,columns=["Country" ,"Total Cases","Total Deaths","Death Percentage"])
df1 
```

```python
# Countries with Highest Death Count per Population
print(":::Top 15 Countries with Highest Death Count:::")
print("-----------------------------------------------")
q2="Select Location, MAX(cast(Total_deaths as int)) as TotalDeathCount from covid_death Where continent<>'' Group by Location order by TotalDeathCount desc limit 15;"
r=cur.execute(q2)
df1=pd.DataFrame(r,columns=["Country" ,"Total Death Couont"])
df1
```

```python
# Getting global numbers
print("::: GLOBAL NUMBERS:::")
print("---------------------")
q1="Select SUM(new_cases) as total_cases, SUM(cast(new_deaths as int)) as total_deaths, SUM(cast(new_deaths as int))/SUM(New_Cases)*100 as DeathPercentage From Covid_Death;"
r=cur.execute(q1)
df1=pd.DataFrame(r,columns=["Total Cases" ,"Total Deaths","Death Percentage"])
df1
```

```python
#Showing contintents with the highest death count per population
print(":::Continents with highest death count:::")
print("------------------------------------------")
q2="Select continent, MAX(cast(Total_deaths as int)) as TotalDeathCount from Covid_death Where continent<>'' Group by continent order by TotalDeathCount desc;"
r=cur.execute(q2)
df1=pd.DataFrame(r,columns=["Continent" ,"Total Death Count"])
df1


```
